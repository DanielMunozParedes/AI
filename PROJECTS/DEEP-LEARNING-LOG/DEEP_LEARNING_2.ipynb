# TIME COMPARISION


```python
import numpy as np
import time
import random
```

```python
n = 1000000

a = [random.random() for i in range(n)]
b = [random.random() for i in range(n)]

s = time.time()
print(s)
```

```python
c = [a[i]*b[i] for i in range(n)]
print("compenhension: ", time.time()-s)

s = time.time()
print("compenhension: ", time.time()-s)

c = []
```

```python
for i in range(n):
  c.append(a[i]*b[i])

print("for loop:", time.time()-s)
s = time.time()
```

```python
c = [0]*n
for i in range(n):
  c[i] = a[i]*b[i]
  
print("existing list:", time.time()-s)
```

```python
x = np.array(a)
y = np.array(b)
s = time.time()
c = x*y
print("NumPy time", time.time()-s)
```
